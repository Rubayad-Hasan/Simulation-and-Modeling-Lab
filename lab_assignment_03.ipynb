Topic: Customer Wait Time Simulation and Analysis
In this lab, we will simulate and analyze customer wait times using both uniform and normal distributions. This helps us model different real-world scenario.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

regular_wait = np.random.uniform(2, 12, 3000)

busy_wait = np.random.normal(8, 2, 3000)

busy_wait = np.where(busy_wait < 0, 0, busy_wait)

print("Data generation complete.")
```

```python
np.random.seed(42)

regular_wait = np.random.uniform(2, 12, 3000)

busy_wait = np.random.normal(8, 2, 3000)

busy_wait = np.where(busy_wait < 0, 0, busy_wait)

print("Data generation complete.")
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(regular_wait, bins=30, color='green', edgecolor='black', alpha=0.9)
plt.title("Regular Days (Uniform Distribution)")
plt.xlabel("Wait Time (minutes)")
plt.ylabel("Frequency")

plt.subplot(1,2,2)
plt.hist(busy_wait, bins=30, color='green', edgecolor='black', alpha=0.9)
plt.title("Busy Days (Normal Distribution)")
plt.xlabel("Wait Time (minutes)")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
