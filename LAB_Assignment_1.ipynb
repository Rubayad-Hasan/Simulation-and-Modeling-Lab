```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#ASSIGNMENT 1:
A = np.random.randint(0,10,(3,3))
B = np.random.randint(0,10,(3,3))
matrix_sum = A+B
matrix_sub = A-B
matrix_mul = A@B
print(matrix_sum)
print(matrix_sub)
print(matrix_mul)
# Matrix Diterminent
det_mata = np.linalg.det(A)
det_matb = np.linalg.det(B)
print(det_mata)
print(det_matb)

# Matrix Inverse for a
if(np.linalg.det(A)!=0):
  inv_mata = np.linalg.inv(A)
else:
  print("Matrix is Singular")

# Matrix Inverse for b
if(np.linalg.det(B)!=0):
  inv_matb = np.linalg.inv(B)
else:
  print("Matrix is Singular")

# Rank of the Matrix
rank_mata = np.linalg.matrix_rank(A)
rank_matb = np.linalg.matrix_rank(B)
print(rank_mata)
print(rank_matb)
```

```python
from matplotlib.lines import lineStyles
A = np.random.uniform(1,50, size=(5,3))
B = np.random.uniform(60,100, size=(5,3))


A_flat=A.flatten()
B_flat=B.flatten()

x =np.arange(len(A_flat))
plt.plot(x,A_flat,marker='o',linestyle='-',label="A")
plt.plot(x,B_flat,marker='s',linestyle='-',label="B")


plt.xlabel("index")
plt.ylabel("values")
plt.title("Line graph of flattened matrices A and B")
plt.legend()
plt.grid(False)

plt.show()



print("Matrix A:\n",A)
print("Matrix B:\n",B)
print("Matrix A flat:\n",A_flat)
print("Matrix B flat:\n",B_flat)
```
