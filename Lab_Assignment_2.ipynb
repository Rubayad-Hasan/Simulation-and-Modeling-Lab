```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d1 = 5
d2 = 8

A = np.array([[d1+7, d2+1],[2*d1, d2+2]])

# Show Matrix
print("Matrix A: \n",A,"\n")
```

```python

size_A = A.shape
det_A = np.linalg.det(A)

rank_A = np.linalg.matrix_rank(A)

eigenvalA = np.linalg.eigvals(A)

if np.linalg.det(A) != 0:
    inv_A = np.linalg.inv(A)
else:
    inv_A = "Matrix is singular, cannot compute inverse"

print("Shape of A:", size_A,"\n")
print("Determinant of A:", det_A,"\n")
print("Rank of A:", rank_A,"\n")
print("Eigenvalues of A:", eigenvalA,"\n")
print("Inverse of A:\n", inv_A,"\n")
```

```python
B = np.array([[d1+2, d2+1],[2*d1, d2+7]])
print("Matrix B:\n",B,"\n")
```

```python
size_B = B.shape
det_B = np.linalg.det(B)
rank_B = np.linalg.matrix_rank(B)

#eigenvalue
eigenvalB = np.linalg.eigvals(B)

if np.linalg.det(B) != 0:
    inv_B = np.linalg.inv(B)
else:
    inv_B = "Matrix is singular, cannot compute inverse"

print("Shape of B:", size_B,"\n")
print("Determinant of B:", det_B,"\n")
print("Rank of B:", rank_B,"\n")
print("Eigenvalues of B:", eigenvalB,"\n")
print("Inverse of B:\n", inv_B,"\n")
```

Q1. How did the determinant change and why?

Changing even one number in a matrix can change the determinant, because the determinant depends on all the numbers inside the matrix.

Q2. Did the rank change? Explain.

Changing one element did not affect the overall structure of the matrix, so the rank stayed the same. Both matrices are still full rank.

Q3. How did the eigenvalues respond to the value change?

The eigenvalues changed significantly. Eigenvalues are derived from the matrix elements, and a change in even one element will generally alter them. They are sensitive to both the trace (sum of diagonal elements) and the determinant of the matrix.

Q4. Is B easier or harder to invert than A? Why?

Both matrices are invertible because their determinants are non-zero. However, Matrix B has a smaller absolute determinant (15.0) compared to Matrix A (30.0). A smaller determinant indicates that the matrix is closer to being singular (non-invertible), which can make its inverse less stable or harder to compute accurately in numerical applications. Therefore, in practical terms, Matrix A is slightly 'easier' (more stable) to invert than Matrix B.
